```python
# 3-4) Show the death rate in Brazil (3 pts)
# Is it above or lower than the worldwide death rate? (3 pts)

number = random.randint(0,10000)
print("You name , id: ", number )
print('------------------------------')

res = df[df['location']=='Brazil']

tot_new_cases = res['new_cases'].sum()

res = df[df['location']=='Brazil']
tot_new_deaths = res['new_deaths'].sum()


rate1 = tot_new_deaths / tot_new_cases
print('Show the death rate in Brazil')
print(format(rate1,'.2%' ))


rate2 = df['new_deaths'].sum()/df['new_cases'].sum()
print('Show the death rate in the world')
print(format(rate2,'.2%' ))

if rate1 > rate2:
    print("Brazil is above the world level")
else:
    print("Brazil is below the world level")
```
